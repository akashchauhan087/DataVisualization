# Matplot lib plot

## Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8.5,9.5]
y = [10,20,30,40,50,65,75,80.5,90.5]
```

```python
plt.plot(x,y)
plt.title('Matplot lib is used')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Scatter

```python
plt.scatter(x,y, color='green')
```

## TITLE xlabel ylabel legend

```python
plt.title('Matplot lib is used')
plt.plot(x,y, color='blue', linestyle=':', marker='o') # marker (o, x)  linestyle '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.scatter(x,y, color='green', marker='x') # marker (o, x)  linestyle '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label="Line 1")
plt.legend()
plt.show()
```

## Bar

```python
plt.bar(['A', 'B', 'C', 'D'], [3, 7, 8, 5], color=['blue', 'green', 'pink', 'yellow'],  width=.2)
plt.show()
```

## Histogram

```python
data = [1, 1, 2, 3, 3, 3, 4, 4, 5]
data2 = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=10, color='purple')
plt.hist(data2, bins=20, color='blue')
plt.show()
```

## Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=(0.1, .15, 0, 0), shadow=True)
plt.show()
```

## Subplot

```python
plt.subplot(2, 1, 1)  # First subplot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('First Plot')

plt.subplot(2, 1, 2)  # Second subplot
plt.bar(['A', 'B', 'C'], [3, 7, 8])
plt.title('Second Plot')

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
plt.subplot(1, 3, 1)  # First subplot
plt.plot([1, 2, 3], [4, 5, 6], color='green')
plt.title('First Plot')

plt.subplot(1, 3, 3)  # Second subplot
plt.bar(['A', 'B', 'C'], [3, 7, 8], color = 'Red')
plt.title('Second Plot')

# plt.tight_layout()  # Adjust spacing
plt.show()
```
